# Frequência Relativa e a Lei Fraca dos Grandes Números

Vamos simular o arremesso de uma moeda confiável e apresentar graficamente a variação da frequência relativa com o número de repetições, verficando a lei fraca dos grandes números.

Pela lei fraca dos grandes números seja $X_1, X_2, \ldots, X_n$ uma sequência de variáveis aleatórias iid com esperança $EX = \mu$, então para $\epsilon > 0$ 

$\lim_{n \to \infty} P[|M_n - \mu| < \epsilon] = 1$,

em que $M_n = (X_1 + X_2 + \ldots + X_n)/n$.

No nosso caso $M_n$ representa a frequência relativa. A lei fraca dos grandes números nos diz que para um número $n$ grande de experimentos a média amostral $M_n$ tende para o valor verdadeiro da esperança com alta probabilidade.

```python
"""
Created on Wed May 27 10:54:03 2020

Simula o arremesso de uma moeda confiável e
apresenta graficamente a variação da frequência relativa
com o número de repetições

@author: albert
"""

import random
import matplotlib.pyplot as plt
import numpy as np

def moeda():
    """
    Simula o arremesso de uma moeda confiável
    10
    Retorna
    -------
    Cara ou Coroa com igual probabilidade
    
    """
    m = round(random.random())
    if(m==0):
        return('Cara')
    else:
        return('Coroa')

def repete(n):
    """
    Repete n vezes o arremesso de uma moeda
    e apresenta as frequências relativas da moeda para cada repetição
    
    Parâmetros
    ----------
    n: número de repetições
    
    obs:
    ------
    frequencia relativa = quantidade de caras (ou coroas)/número de arremessos
    ------
    """
    ncaras = ncoroas = 0
    fcaras = []
    fcoroas = []
    for i in range(n):
        fcaras.append(0) 
        fcoroas.append(0)
    for j in range(n):
       arremesso = moeda()
       
       if(arremesso == 'Cara'):
           ncaras += 1
       else:
           ncoroas += 1

        ### guarda as frequências relativas parciais 
       fcaras[j] = ncaras/(j+1)
       fcoroas[j] = ncoroas/(j+1)

    ### Apresenta as frequências relativas finais     
    print("Frequência Relativa de Cara ", ncaras/n)
    print("Frequência Relativa de Coroas ", ncoroas/n)
#   print(fcaras)
#   print(fcoroas)
    
    ###  Mostra graficamente
    x = np.arange(1,n+1,1)
    plt.figure(figsize=(10,5),dpi=70)
    plt.plot(x,fcaras,label='cara')
    plt.plot(x,fcoroas,label='coroa')
    plt.title('Convergência da Freq. Relativa')
    plt.ylabel('Freq. Relativa')
    plt.xlabel('n')
    plt.minorticks_on()
    plt.grid(b=True, which='minor', color='#999999', linestyle='-', alpha=0.2)
    plt.legend()    
    plt.show()
    
    
def menu():
    """
    Menu de entrada do número de repetições
    
    """
    
    n = int(input('Quantos arremessos da moeda? '))
    repete (n)
    
while True:
    menu()
```
