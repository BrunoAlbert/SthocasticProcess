# Kalman exemplo


Modelo (escalar): \begin{align} Y[n] &= aY[n-1] + W[n] \\ X[n] &= cY[n] + N[n] \end{align}
                  
em que $W[n]$ e $N[n]$ gaussianas
                  
Saída: $y[n]$ estimativa de $Y[n]$

Inicialização:
    
$EW[n] = EN[n] = 0$, $EW^2 = 0,36$, $Ee^2[0] = 0$, $EN^2 = 1$, $Y[0] = 0$, $n = 100$, $a = 0,8$ e $c = 1$

```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Mon May 24 17:32:53 2021

@author: albert
IQuanta - DEE - UFCG
Data: maio 2021
Versão: 1.0

Modelo (escalar): Y[n] = aY[n-1] + W[n]
                  X[n] = cY[n] + N[n]
                  
                  W[n] e N[n] gaussianas
                  
Saída: y[n] estimativa de Y[n]

Inicialização:
    
    EW[n] = EN[n] = 0, EW**2 = 0,36, Ee[0]**2 = 0, EN**2 = 1, Y[0] = 0, n = 100, a = 0,8
"""

import random
import matplotlib.pyplot as plt
import numpy as np
import math

### Inicialização
n = 100 # número de amostras
EW = EN = 0 # Esperança de W e N
EW2 = 0.36  # variância de W[n]
EN2 = 1.0 # variância do ruído N[n]
a = 0.8 # parâmetro do modelo
c = 1   # parâmetro do modelo


W = []  # vetor W[n]
N = []  # vetor N[n]
for i in range(n):
   W.append(random.gauss(EW, math.sqrt(EW2)))
   N.append(random.gauss(EN, math.sqrt(EN2)))
     
### vetore Y e X
Y = []
X = []
for k in range(n):
    Y.append(0)
    X.append(0)
    
for i in range(1,n):
    Y[i] = a*Y[i-1] + W[i]
    X[i] = c*Y[i] + N[i]
    
###  Mostra graficamente
    
x = np.arange(1,n+1,1)
plt.figure(figsize=(12, 10))
plt.plot(x,Y,label='Y')
plt.plot(x,X,label='X')
plt.title('Y e X')
#   plt.ylabel('Freq. Relativa')
#   plt.xlabel('n')
plt.show()
```

Estimativa do filtro de Kalman

Gráfico do ganho

```python
## Estimativa de Kalman
y = []  # estimativa de Y[n]
k = []  # ganho
Ee2 = []    # erro quadrático
for i in range(n):
    y.append(0)
    k.append(0)
    Ee2.append(0)
    
y.append(0)
Ee2.append(0)

for i in range(n):
    k[i] = a*Ee2[i]/(Ee2[i] + EN2)
    y[i+1] = a*y[i] + k[i]*(X[i] - y[i])
    Ee2[i+1] = a*(a - k[i])*(Ee2[i]) + EW2

# gráfico do ganho
plt.figure(figsize=(12, 10))
plt.plot(x,k)
plt.title('ganho')
plt.show()
```

Saída do filtro de Kalman

```python
# gráfico da estimativa do filtro de Kalman
plt.figure( figsize=(12, 10)) 
plt.plot(x,Y,label='Y')
plt.plot(x,X,label='X')
x = np.arange(0,n+1,1)
plt.plot(x,y,label='y')
plt.title('Y, X e y')
# plt.figure( figsize=(40, 18)) 
plt.show()
```
