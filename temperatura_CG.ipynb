# Correlação de Temperaturas

Tomamos a temperatura de Campina Grande em março de 2015 às 6, 11 e 12 horas, fonte INMET.

Pretende-se calcular esperanças, desvios padrão, e correlações e também mostrar os histogramas e os diagramas de espalhamento dos 3 horários acima.

## Carregar as temperaturas

```python
"""
Created on Mon Jul 19 10:55:44 2021

@author: albert
"""
import statistics as st
import random
import matplotlib.pyplot as plt
import numpy as np
import math

# abrir os arquivos

ft6 = open("temperatura_06h_03_2015")
ft11 = open("temperatura_11h_03_2015")
ft12 = open("temperatura_12h_03_2015")

t6 = ft6.readlines()
t11 = ft11.readlines()
t12 = ft12.readlines()

T6 = list(map(lambda var: float(var), t6))
T11 = list(map(lambda var: float(var), t11))
T12 = list(map(lambda var: float(var), t12))

print("Temperaturas às 6 h em CG mar/2015 ", T6, "\n")
print("Temperaturas às 11 h em CG mar/2015 ", T11, "\n")
print("Temperaturas às 12 h em CG mar/2015 ", T12, "\n")
```

## Histogramas das temperaturas

Às 6 horas

```python
plt.figure(figsize=(8,6))
plt.title("Histograma 6 horas, mar 2015")
plt.ylabel("", fontsize=20)
plt.xlabel("Temperaturas", fontsize=20)
n, bins, patches = plt.hist(T6, bins=5)
```

Às 11 horas

```python
plt.figure(figsize=(8,6))
plt.title("Histograma 11 horas, mar 2015")
plt.ylabel("", fontsize=20)
plt.xlabel("Temperaturas", fontsize=20)
n, bins, patches = plt.hist(T11, bins=5)
```

Às 12 horas

```python
plt.figure(figsize=(8,6))
plt.title("Histograma 12 horas, mar 2015")
plt.ylabel("", fontsize=20)
plt.xlabel("Temperaturas", fontsize=20)
n, bins, patches = plt.hist(T12, bins=5)
```

## Esperanças

```python
print("esperanças: \n6h - ", st.mean(T6), "\n11h - ", st.mean(T11), "\n12h - ", st.mean(T12))
```

## Desvios padrão

```python
print("desvios padrão: \n6h - ", st.stdev(T6), "\n11h - ", st.stdev(T11), "\n12h - ", st.stdev(T12))
```

## Diagramas de espalhamento

6 horas X 12 horas

```python
plt.figure(figsize=(8,8))
plt.title("Diagrama de espalhamento", fontsize=20)
plt.ylabel("12 horas", fontsize=20)
plt.xlabel("6 horas", fontsize=20)
plt.scatter(T6,T12)
plt.show()
```

```python
plt.figure(figsize=(8,8))
plt.title("Diagrama de espalhamento", fontsize=20)
plt.ylabel("11 horas", fontsize=20)
plt.xlabel("6 horas", fontsize=20)
plt.scatter(T6,T11)
plt.show()
```

11 horas X 12 horas

```python
plt.figure(figsize=(8,8))
plt.title("Diagrama de espalhamento", fontsize=20)
plt.ylabel("12 horas", fontsize=20)
plt.xlabel("11 horas", fontsize=20)
plt.scatter(T11,T12)
plt.show()
```

## Coeficientes de correlação

$\rho(6,12)$

```python
r6_12 = np.corrcoef(T6,T12)
print("coeficiente de correlação: 6h - 12h - ", r6_12[0,1])
```

$\rho(6,11)$

```python
r6_11 = np.corrcoef(T6,T11)
print("coeficiente de correlação: 6h - 11h - ", r6_11[0,1])
```

$\rho(11,12)$

```python
r11_12 = np.corrcoef(T11,T12)
print("\ncoeficiente de correlação: 116h - 12h - ", r11_12[0,1])
```

## Correlação temperatura entre CG e JP

Vamos calcular a correlaçao das temperaturas entre CG e JP em janeiro de 2016 ao meio dia

```python
cg12 = open("temperatura_CG_12h_01_2016.dat")
jp12 = open("temperatura_jp_12h_01_2016.dat")

tcg12 = cg12.readlines()
tjp12 = jp12.readlines()

fcg12 = list(map(lambda var: float(var), tcg12))
fjp12 = list(map(lambda var: float(var), tjp12))
```

Esperanças e desvios padrão

```python
print("esperanças: \nCG 12h - ", st.mean(fcg12), "\nJP 12h - ", st.mean(fjp12))
print("desvios padrão: \nCG 12h - ", st.stdev(fcg12), "\nJP 12 h - ", st.stdev(fjp12))
```

Diagrama de espalhamento

```python
plt.figure(figsize=(8,8))
plt.title("Diagrama de espalhamnento", fontsize=20)
plt.ylabel("12 horas JP", fontsize=20)
plt.xlabel("12 horas CG", fontsize=20)
plt.scatter(fcg12,fjp12)
plt.show()
```

$\rho_T(CG12,JP12)$

```python
rcg_jp_12 = np.corrcoef(fcg12,fjp12)

print("coeficiente de correlação: CG - JP - 12h - ", rcg_jp_12[0,1])
```
