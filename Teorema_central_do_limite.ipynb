# Teorema Central do Limite

Seja $S_n=X_1+X_2+\ldots+X_n$ a soma de $n$ v.as. iid com esperança finita $EX_i=\mu$ e variância finita $\text{var}(X_i)=\sigma²$. Defina $Z_n$ como uma v.a. com esperança zero e variância unitária da seguinte forma

$Z_n = \frac{S_n - n\mu}{\sigma\sqrt{n}}$
     
então

$\lim_{n\to \infty}P[Z_n \le z] = \frac{1}{\sqrt{2\pi}} \int_{-\infty}^z e^{-x²/2} \,dx.$

Ou seja, a distribuição da v.a. $S_n$ tende para uma gaussiana quando $n \to \infty$

Na prática valores mais modestos de $n$ já produzem resultados que se aproximam bastante de uma distribução gaussiana.

## Programa para demonstrar o teorema central do limite

O programa a seguir demonstra o teorema central do limite. Vamos determinar a função densidade de probabilidade da soma $S_n$ de $n$ v.as. iid $X_i, \quad i = 1,2,\ldots,n$. 

$X_i$ tem uma distribuição uniforme no intervalo $[0,1)$

$S_n$ tem uma distribuição implementada por $n-1$ sucessivas convoluções que é mostrada graficamente.

### FDP de X

```python
"""
@author: albert
IQuanta - DEE - UFCG
Data: outubro 2022
Versão: 1.0
"""

import matplotlib.pyplot as plt
import numpy as np
from numpy import random

"""
Demonstração do Teorema Central do Limite

Sn = X1 + X2 + ... + Xn

fX(x) - fdp de Xi, uniforme no intervalo [0,1)

fSn(x)= fX1(x) * fx2(x) * ... * fXn(x) - fdp de Sn (*, convolução) 
"""

# fdp de X
passo = 0.005
x = np.arange(0,1,passo)
fx = np.ones(len(x))

plt.figure(figsize=(16, 8))
plt.plot(x,fx,label='fx(x)')
# plt.legend()
plt.title('fdp de X')
plt.xlabel('x')
plt.show()
```

### FDP de $S_2$

```python
S2 = np.convolve(fx, fx)
print(len(S2))

x2 = np.arange(0,2- passo,passo)
print(len(x2))
plt.figure(figsize=(16, 8))
plt.plot(x2,S2,label='S2')
plt.legend()
plt.title('fdp de S2')
plt.xlabel('x')
plt.show()
```

### FDP de $S_3$

```python
S3 = np.convolve(S2, fx)
print(len(S3))

x3 = np.arange(0,3-2*passo,passo)
print(len(x3))
plt.figure(figsize=(16, 8))
plt.plot(x3,S3,label='S3')
plt.legend()
plt.title('fdp de S3')
plt.xlabel('x')
plt.show()
```

### FDP de $S_4$

```python
S4 = np.convolve(S3, fx)
print(len(S4))

x4 = np.arange(0,4-3*passo,passo)
print(len(x4))
plt.figure(figsize=(16, 8))
plt.plot(x4,S4,label='S4')
plt.legend()
plt.title('S4(x)')
plt.xlabel('x')
plt.show()
```
